```python
import p as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel('ticket.xlsx')
```

```python
df
```

```python
x= df['PaymentMode']
y= df['TicketPrice']
plt.bar(x,y)
plt.bar(df['TicketType'],y)
```

```python
data = df.groupby('PaymentMode')['TicketPrice'].sum()
```

```python
data
```

```python
data =data.reset_index()
```

```python
data
```

```python
colors=['red','blue','green','pink']
```

```python
plt.bar(data['PaymentMode'],data['TicketPrice'],color=colors,edgecolor='black',linewidth=1.8,label=data['PaymentMode'])
plt.legend()
```
